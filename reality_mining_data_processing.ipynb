```python
dic_lab = pd.DataFrame({},columns=[0,1,2,'label'])
for x in loc_seq.keys():
    for i in range(0,22):
        df = loc_seq[x][range(i,i+3)]
        if i > -1 and i<7:
            l = 1
        if i >= 7 and i<9:
            l = 2
        if i >= 9 and i<11:
            l = 3
        if i >= 11 and i<14:
            l = 4
        if i >= 14 and i<17:
            l = 5
        if i >= 17 and i<19:
            l = 6
        if i >= 19 and i<21:
            l = 7
        if i >= 21 and i<24:
            l = 8
        df.columns = [0,1,2]
        df['label'] = pd.Series([`l`]*len(df))
        df['person'] = pd.Series([x]*len(df))
        dic_lab = dic_lab.append(df)
    
dic_lab=dic_lab.reset_index().rename(columns = {'index':'day'})
```

```python
map1 = {0:'N', 1:'H', 2:'W', 3:'O', np.nan: 'X'}
dic_lab[0] = dic_lab[0].map(map1)
dic_lab[1] = dic_lab[1].map(map1)
dic_lab[2] = dic_lab[2].map(map1)
loc_labs = dic_lab[[0,1,2,'label']]
loc_labs['count'] = [1]*len(loc_labs)
```

```python
dat_corp = {}
for p in set(dic_lab.person.values):
    dat_corp[p]={}
    for d in set(dic_lab.day.values):
        df = dic_lab[(dic_lab['person']==p) & (dic_lab['day']==d)][[0,1,2,'label']]
        df_word = pd.DataFrame(df[0]+df[1]+df[2]+df['label'], columns=['day_seq']).fillna('xxxx')
        dat_corp[p][d]= ' '.join(df_word.day_seq.values)
```
